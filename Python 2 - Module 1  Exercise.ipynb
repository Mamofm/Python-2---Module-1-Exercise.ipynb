## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

````python
class Rectangle():
    """ Will create a Rectangle class with a height and length attributes.  
    Class will contain two methods calculating preimeter() and area(). """
    def __init__ (self, height, length):
        self.height = height
        self.length = length

    def perimeter (self):
        p = self.height+ self.height + self.length + self.length 
        return p
    
    def area (self):
        a = int(self.height) + int(self.length)
        return a

    
# main 
rec1 = Rectangle(12, 3)  # rectangle 1
rec2 = Rectangle(4, 5)  # rectangle 2


#Compare two triangle's preimeter and print result 

if (rec1.perimeter() > rec2.perimeter()):
    print (f"1st rectangle perimeter is {rec1.perimeter()} and 2nd rectangle perimeter is {rec2.perimeter()}")
    print (f"1st rectangle perimeter is larger than 2nd rectangle")
else:
    print (f"2nd rectangle perimeter is {rec2.perimeter()} and 1st rectangle perimeter is {rec1.perimeter()}")
    print (f"2nd rectangle perimeter is larger than 1st rectangle")

````

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



````python
import math
class Triangle():
    """ Will create a Triangle class with a height and base attributes.  
    Class will contain two methods calculating preimeter() and area(). """
    def __init__ (self, height, base):
        self.height = height
        self.base = base

    def perimeter (self):
        h1 = self.height * self.height
        b1 = self.base * self.base
        c = math.sqrt( b1 + h1 )
        p = self.height + self.base + c
        return p
    
    def area (self):
        a = (float(self.height) * float(self.base)) / 2
        return a

    
# main 
tri1 = Triangle (12,3)  #triangle 1
tri2 = Triangle (4, 5)  # triangle 2

#Compare two triangle's preimeter and print result 
if (tri1.perimeter() > tri2.perimeter()):
    print (f"1st triangle perimeter is {tri1.perimeter():,.2f} and 2nd triangle perimeter is {tri2.perimeter():,.2f}")
    print (f"1st triangle perimeter is larger than 2nd triangle")
else:
    print (f"2nd triangle perimeter is {tri2.perimeter():,.2f} and 1st triangle perimeter is {tri1.perimeter():,.2f}")
    print (f"2nd triangle perimeter is larger than 1st triangle")

````
